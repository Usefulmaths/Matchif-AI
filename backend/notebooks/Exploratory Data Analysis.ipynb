```python
import pandas as pd

# Load the dataset
df = pd.read_csv('data/postings.csv')

# Display the first few rows of the dataframe
df.head()
```

```python
# Get a summary of the dataframe
df.info()
```

```python
# Get descriptive statistics
df.describe(include='all')
```

```python
# Check for duplicates
duplicates = df.duplicated().sum()

# Print the number of duplicates found
print(f"Number of duplicate entries found: {duplicates}")

# If duplicates are found, remove them and display the new count
if duplicates > 0:
    df = df.drop_duplicates()
    print(f"Duplicates removed. New number of entries: {len(df)}")
```

```python
# Select columns that are most relevant for web application
selected_columns = [
    'title', 'company_name', 'description', 'location',
    'max_salary', 'min_salary', 'currency', 'pay_period',
    'formatted_work_type', 'remote_allowed',
    'application_url', 'application_type', 'formatted_experience_level'
]

# Create a subset dataframe with the selected columns
df_selected = df[selected_columns]

df_selected.info()
```

```python
# Function to display unique values in each column
def display_unique_values(df, columns):
    for column in columns:
        unique_values = df[column].unique()
        num_unique_values = len(unique_values)
        if num_unique_values < 10:  # Display only if unique values are less than 50
            print(f'\nColumn: {column} (Unique Values: {num_unique_values})')
            print(unique_values)
        else:
            print(f'\nColumn: {column} (Unique Values: {num_unique_values})')
            print(f'Too many unique values to display.')

# Display unique values for each selected column
display_unique_values(df_selected, selected_columns)
```

```python
# Drop rows with missing values in critical columns
critical_columns = ['title', 'company_name', 'description', 'location', 'application_url']
df_selected = df_selected.dropna(subset=critical_columns)

# Impute missing values for other columns
df_selected['remote_allowed'] = df_selected['remote_allowed'].fillna(0)
df_selected['formatted_experience_level'] = df_selected['formatted_experience_level'].fillna('Unknown')

# Verify the changes
df_selected.info()
```

```python
import re

# Function to check for unusual spacing in descriptions
def check_unusual_spacing(description):
    return bool(re.search(r'\s{2,}', description)) or bool(re.search(r'\n+', description))

# Apply the check function to the description column
df_selected['unusual_spacing'] = df_selected['description'].apply(check_unusual_spacing)

# Count the number of descriptions with unusual spacing
num_unusual_spacing = df_selected['unusual_spacing'].sum()

# Display the result
print(f"Number of descriptions with unusual spacing: {num_unusual_spacing}")

# Display a sample of descriptions with unusual spacing
descriptions_with_issues = df_selected[df_selected['unusual_spacing']]

descriptions_with_issues[['description']].head(10) # Display the first 10 for a quick check
```

```python
# Function to clean description text
def clean_description(description):
    # Remove HTML tags
    description = re.sub(r'<.*?>', '', description)
    # Replace newlines with full stops
    description = description.replace('\n', '. ')
    # Normalise spacing
    description = re.sub(r'\s+', ' ', description).strip()
    return description

# Apply the cleaning function to the description column
df_selected['description'] = df_selected['description'].apply(clean_description)

# Verify the changes
df_selected.info()
```
